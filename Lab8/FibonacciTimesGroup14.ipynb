```python
# fibR - Fibonacci Recursive O( ( (1+sqrt(5)/2 )^n lg n ) [Relaxed asymptotic bound O(2^n lg n)] 
def fibR(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibR(n-1)+fibR(n-2)    # for large nÂ´s  the addition will take lg n stetp because the numebers are very large
```

```python
from time import process_time

def time_recursive_fib(n):
  t0 = process_time()
  fibR(n)
  tf =process_time()
  return tf-t0
time_recursive_fib(10)
```

```python
import math
import matplotlib.pyplot as plt
timeList = [0]
n = 42
for i in range(1,n):
  timeList.append(time_recursive_fib(i))

plt.plot(range(1,n),timeList[1:])
plt.plot(range(1,n),range(1,n))
```

```python
# fibI - Fibonacci Iterartive (forward) O(n lgn)
def fibI(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```python
from time import process_time

def time_recursive_fib(n):
  t0 = process_time()
  fibI(n)
  tf =process_time()
  return tf-t0
```

```python
import math
import matplotlib.pyplot as plt
timeList = [0]
n = 1001
for i in range(1,n):
  timeList.append(time_recursive_fib(i))

plt.plot(range(1,n),timeList[1:])
plt.plot(range(1,n),range(1,n))
```
